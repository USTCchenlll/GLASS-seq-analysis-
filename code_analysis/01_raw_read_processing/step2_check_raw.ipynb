```python
import gzip
import os

# ==== 参数 ====
# 只输入样本名（不带后缀）
samples = ["1a","1b","1s","2a","2b","2s","3a","3b","3s",]

flash_path = "D:/CL/AFISH1/afish_data_20251009/flash_data/"
suffix = ".extendedFrags.fastq"   # 所有文件共有的后缀
minlen, maxlen = 140, 160 # 长度范围

# ==== 主循环 ====
for name in samples:
    in_path = flash_path + name + suffix
    out_path = flash_path + name + "_filtered" + suffix

    if not os.path.exists(in_path):
        print(f"⚠️ 找不到文件: {in_path}")
        continue

    open_in = gzip.open if in_path.endswith('.gz') else open
    open_out = gzip.open if out_path.endswith('.gz') else open

    total = kept = short = long = 0

    with open_in(in_path, 'rt', encoding='utf-8') as fin, open_out(out_path, 'wt', encoding='utf-8') as fout:
        while True:
            header = fin.readline()
            if not header:
                break
            seq = fin.readline()
            plus = fin.readline()
            qual = fin.readline()
            if not qual:
                break

            total += 1
            L = len(seq.strip())
            if L < minlen:
                short += 1
            elif L > maxlen:
                long += 1
            else:
                kept += 1
                fout.write(header)
                fout.write(seq)
                fout.write(plus)
                fout.write(qual)

    print(f"\n✅ 处理完成: {in_path}")
    print(f"  总序列数: {total}")
    print(f"  过短 (<{minlen}): {short}")
    print(f"  过长 (>{maxlen}): {long}")
    print(f"  保留 ({minlen}–{maxlen}): {kept} ({kept/total*100 if total else 0:.2f}%)")
    print(f"  输出文件: {out_path}")
```
